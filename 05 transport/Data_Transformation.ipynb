```python
import pandas as pd
from datetime import datetime, timedelta

file_path = 'bc_230215.csv'

columns_to_keep = [col for col in pd.read_csv(file_path, nrows=1).columns if col not in ['EVENT_NO_STOP', 'GPS_SATELLITES', 'GPS_HDOP']]
df_filtered = pd.read_csv(file_path, usecols=columns_to_keep)

def decode_timestamp(row, opd_date_col='OPD_DATE', act_time_col='ACT_TIME'):
    opd_date_str = row[opd_date_col]
    opd_date = datetime.strptime(opd_date_str[:-9], '%d%b%Y')
    act_time = timedelta(seconds=row[act_time_col])
    return opd_date + act_time

df_filtered['TIMESTAMP'] = df_filtered.apply(decode_timestamp, axis=1)

grouped = df_filtered.groupby('VEHICLE_ID')

for name, group in grouped:
    max_speed = 0
    max_speed_trip = None
    max_speed_location = None
    max_speed_time = None
    median_speeds = []

    unique_trips = group['EVENT_NO_TRIP'].unique()

    for trip in unique_trips:
        trip_data = group[group['EVENT_NO_TRIP'] == trip].copy()

        trip_data = trip_data.drop(['OPD_DATE', 'ACT_TIME'], axis=1)

        trip_data['dMETERS'] = trip_data['METERS'].diff()
        trip_data['dTIMESTAMP'] = trip_data['TIMESTAMP'].diff()

        trip_data['SPEED'] = trip_data.apply(lambda row: row['dMETERS'] / row['dTIMESTAMP'].total_seconds() if row['dTIMESTAMP'].total_seconds() != 0 else 0, axis=1)

        trip_data = trip_data.drop(['dMETERS', 'dTIMESTAMP'], axis=1)

        trip_max_speed = trip_data['SPEED'].max()
        if trip_max_speed > max_speed:
            max_speed = trip_max_speed
            max_speed_trip = trip
            max_speed_row = trip_data.loc[trip_data['SPEED'].idxmax()]
            max_speed_location = (max_speed_row['GPS_LATITUDE'], max_speed_row['GPS_LONGITUDE'])
            max_speed_time = max_speed_row['TIMESTAMP']
        median_speed = trip_data['SPEED'].median()
        median_speeds.append(median_speed)

    overall_median_speed = pd.Series(median_speeds).median()

    print(f"Vehicle ID: {name}")
    print(f"Maximum Speed: {max_speed} m/s")
    print(f"Trip: {max_speed_trip}")
    print(f"Location: {max_speed_location}")
    print(f"Time: {max_speed_time}")
    print(f"Overall Median Speed: {overall_median_speed} m/s")
    print()
```
